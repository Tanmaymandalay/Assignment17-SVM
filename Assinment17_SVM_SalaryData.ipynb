```python
# SVM Classification
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report


from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score

import warnings as wn
wn.filterwarnings('ignore')
```

```python
sal_data=pd.read_csv('/content/SalaryData.csv')
sal_data.head()
```

```python
sal_data.shape
```

```python
sal_data['Salary'].unique()
```

```python
sal_data1=pd.get_dummies(sal_data.iloc[:,0:13])
```

```python
sal_data1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
sal_data['Salary']=le.fit_transform(sal_data['Salary'])
salary=sal_data['Salary']
```

```python
sal_data1=sal_data1.join(salary)
```

```python
X=sal_data1.iloc[:,0:98]
Y=sal_data1.iloc[:,-1]
```

```python
sc=StandardScaler()
X_std=sc.fit_transform(X)
```

```python
X_std,Y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_std,Y, test_size = 0.3)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
model = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[50,5,10,0.5],'C':[15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(model,param_grid,cv=3)
```

```python
gsv.fit(X_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
model = SVC(C= 13, gamma = 0.5)
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
